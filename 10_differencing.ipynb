## Differencing

```python
from sktime.transformations.series.difference import Differencer
from sktime.utils.plotting import plot_series
from sktime.datasets import load_airline
```

```python
y = load_airline()
```

```python
plot_series(y)
```

```python
from matplotlib import pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(y, lags=50)
```

```python
transformer = Differencer(lags=[1])
y_transform_1 = transformer.fit_transform(y)
```

```python
plot_series(y_transform_1)
```

```python
plot_acf(y_transform_1, lags=50)
```

```python
transformer = Differencer(lags=[12])
y_transform_12 = transformer.fit_transform(y)
plot_series(y_transform_12)
```

```python
plot_acf(y_transform_12)
```

```python
transformer = Differencer(lags=[1, 12])
y_transform_1_12 = transformer.fit_transform(y)
plot_series(y_transform_1_12)
```

```python
plot_acf(y_transform_1_12)
```
